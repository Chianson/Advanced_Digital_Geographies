In this assignment we scraped the UW geography faculty website and created a csv file with names of professors, their titles, and their email address.  Kuo came up with the CSS selections for data we needed and typed up the code, and the rest contributed by debugging and coming up with the way to write to csv.


```python
import urllib2
import lxml
from lxml import html
import pandas
import os
```

```python
# Stores geography faculty website to local file
u = urllib2.urlopen('https://geography.washington.edu/people/faculty')
localFile = open('data/file.html', 'w')
localFile.write(u.read())
localFile.close()
```

```python
# Reads the local html file
path = os.path.abspath("data/file.html")
doc_text = urllib2.urlopen("file:///" + path).read()
doc = html.fromstring(doc_text)
```

```python
data = []

# Get the first element with .view-content because we're only interested in the first div with this class
ele = doc.cssselect(".view-content")[0]

# For every div inside the selected element
for row in ele.cssselect(".view-content > div"):
    name = row.cssselect("h3 a").pop().text # Get professor name from anchor in header
    title = row.cssselect(".views-field-field-jobtitle > div").pop().text # Get title from div within div
    email = row.cssselect(".views-field-field-email > div a").pop().text # Get email from anchor within div
    to_append = [name, title, email]
    data.append(to_append)

# Add column name and convert data to a csv file
pandas_df = pandas.DataFrame(data,columns=['name','title','email'])
pandas_df.to_csv("data/faculty.csv", index = False, encoding='utf-8')
pandas_df
```
